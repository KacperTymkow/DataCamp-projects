# Project Overview: Analyzing Students' Mental Health

## Introduction
Does going to university in a different country affect your mental health? In 2018, a Japanese international university surveyed its students regarding their mental well-being and published a comprehensive study the following year. 

The published research concluded that international students face a higher risk of mental health difficulties than the general population. Furthermore, the study identified that **social connectedness** (belonging to a social group) and **acculturative stress** (stress associated with adapting to a new culture) are strong predictors of depression.

## Project Objective
The goal of this project is to leverage **SQL** to explore the university's dataset and verify if the data aligns with the published academic conclusions. Specifically, this analysis focuses on investigating whether the **length of stay** in a foreign country is a contributing factor to the international students' mental health challenges.



## Data Description
Below is a description of the key columns used in this analysis:

| Field Name | Description |
| :--- | :--- |
| `inter_dom` | Types of students (international or domestic) |
| `japanese_cate` | Japanese language proficiency |
| `english_cate` | English language proficiency |
| `academic` | Current academic level (undergraduate or graduate) |
| `age` | Current age of student |
| `stay` | Current length of stay in years |
| `todep` | Total score of depression (PHQ-9 test) |
| `tosc` | Total score of social connectedeness (SCS test) |
| `toas` | Total score of acculturative stress (ASISS test) |

```python
%load_ext sql
%sql duckdb://
```

```python
%%sql
CREATE OR REPLACE VIEW students AS SELECT * FROM 'students.csv';
```

```python
%%sql
SELECT stay, COUNT(*) AS count_int,
    ROUND(AVG(todep),2) AS average_phq, 
    ROUND(AVG(tosc),2) AS average_scs,
    ROUND(AVG(toas),2) AS average_asiss
FROM students
WHERE Inter_dom = 'Inter'
GROUP BY stay
ORDER BY stay DESC;
```
